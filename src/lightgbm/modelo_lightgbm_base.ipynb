# Prueba con LighGBM

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import lightgbm as lgb
```

```python
#EXPERIMENTO
experimento = 'EN-001'

#CARGO DATASETS
os.chdir("C:/Users/vyago/Desktop/Yago/Competencia/ypf")  # Directorio actual
train = pd.read_csv("../Dataset/dataset_train.csv")
test = pd.read_csv("../Dataset/dataset_test.csv")
```

```python
train = train.select_dtypes("number")
```

```python
y_train = np.sqrt(train["delta_WHP"])
x_train = train[train.columns.drop(["delta_WHP","ID_FILA"])]#,"ID_EVENTO"])]
```

```python
train_data = lgb.Dataset(x_train, label=y_train)
```

```python
# Parámetros 

param = {'max_bin':256,
         'objective': 'regression',
         'learning_rate':0.1796,
         'num_iterations':250,
         'min_data_in_leaf':2100,
         'feature_fraction':0.2053,
         'num_leaves':1983}

param['metric'] = 'rmse' # métrica
```

```python
modelo=lgb.train(param,train_data)
```

```python
x_test = test[x_train.columns]
x_test = x_test.astype("float32")
test["delta_WHP"]=np.exp(modelo.predict(x_test))
```

```python
prediccion=test[["ID_FILA","delta_WHP"]]
prediccion.to_csv("../Exp/base_sqr.csv",header=False,index=False)
```

```python
if not os.path.isdir(f'../Exp/{experimento}'):
    os.makedirs(f'../Exp/{experimento}')
 
prediccion.to_csv(f"../Exp/{experimento}/prediccion.csv",header=False,index=False)
```
